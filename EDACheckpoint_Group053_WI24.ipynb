# Research Question

What factors contribute to the density and migration patterns of monarch butterflies over time?


## Background and Prior Work

This brief literature review will examine prior work relating to the migration patterns of Monarch butterflies (*Danaus plexippus*) and possible catalysts. Our research seeks to elucidate factors that contribute to the migration patterns and population density of Monarch butterflies over time. Studies such as Lemoine (2015) suggest that climate change is likely to alter the breeding ground distributions of eastern migratory Monarchs via range expansion of Asclepias host plants, potentially affecting Monarch distributions due to changes in host plant availability and habitat sustainability <a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1). Continuing on the climate front, Cohen (2023) highlighted that Monarchs born under autumn-like conditions showed lower post-metabolic rates and greater flight efficiency compared to those born under summer-like conditions <a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2). This observation coupled with historic climate data could aid in determining causal factors for migration pattern anomalies present in the population data. Climate aside, Flockhart et al (2015) have demonstrated that habitat loss on breeding grounds due to agricultural practices has been identified as a significant factor driving population declines in Monarch butterflies <a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3). Monarch butterflies utilize milkweed in cornfields throughout their breeding season, Oberhauser et al (2001) suggests that due to per-plant densities being higher in agricultural habitats than nonagricultural habitats, coupled with the prevalence of agricultural land, most Monarchs produced in the upper Midwest originated from cornfields or another agricultural land <a name="cite_ref-4"></a>[<sup>4</sup>](#cite_note-4). The researchers suggest that agricultural practices such as weed control and insecticide use could significantly impact Monarch populations due to the loss of milkweed habitats in the areas aforementioned. While the studies by Lemoine (2015) and Flockhart et al. (2015) offer significant insights into the impact of climate change and habitat loss on Monarch populations, there is a gap in our understanding of the relationship between climate events and migration timing; we seek to bridge this gap by employing statistical techniques and machine learning. Methods to track the migration patterns and population density follow. 

Mehmood et al (2023) compiled a project to track Monarch butterfly migration patterns utilizing web scraping, data processing, mapping via Python, and a Google Cloud AutoML Image Classification Model<a name="cite_ref-5"></a>[<sup>5</sup>](#cite_note-5). With these methods the researchers aggregated Citizen Science data related to Monarch butterfly sightings and verified said sightings using machine learning. Similarly, our project will utilize Citizen Science coupled with observations from iNaturalist to refine the analysis of migration patterns. Monarch sighting and temperature data were categorized from 1997-2023 and plotted on a map of the US(link to map). Recent studies have suggested that Monarchs have experienced an alarming decline in population size due to a combination of deforestation, loss of native milkweed and nectar plants, and climate change. In a similar approach, Patel et al (2021) developed an open-source tool utilizing deep convolutional neural networks to track Monarch sightings <a name="cite_ref-6"></a>[<sup>6</sup>](#cite_note-6). Mahr (2024) developed a visualization project in R that visualizes the annual migration patterns of Monarch butterflies using data from the Global Biodiversity Information Facility <a name="cite_ref-7"></a>[<sup>7</sup>](#cite_note-7). This project showcases the movement patterns of Monarchs month-over-month providing a strong basis from which we may architect our code for similar migration analysis. By adapting these techniques, we aim to further our understanding of Monarch migration patterns about habitat and climate change. These studies make explicit the ability to track the migration patterns of Monarch butterflies synchronously with variables such as temperature and deforestation. 

Our project aims to build upon the work of the aforementioned studies, employing machine learning techniques to accurately predict Monarch migration patterns in response to climate change, milkweed density, and other confounding variables that may be uncovered throughout the course of this research.

```python
# imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import xarray as xr
from datetime import datetime, timedelta
import folium
from folium.plugins import HeatMap
from sklearn.cluster import KMeans
import seaborn as sns
```

## Inaturalist data (Monarch and Milkweed)

```python
#monarch butterfly csv files from iNaturalist
monarch_2008_2021 = pd.read_csv('monarch_2008-2021.csv')
monarch_2021_2024 = pd.read_csv('monarch_2021-2024.csv')
monarch_df = pd.concat([monarch_2008_2021, monarch_2021_2024], ignore_index=True)

#cleaning monarch data
columns_to_keep = ['observed_on', 'latitude', 'longitude']
monarch_df = monarch_df[columns_to_keep]

# drop null rows
monarch_df = monarch_df.dropna()

# rename observed_on column to time
monarch_df.rename(columns={'observed_on': 'time'}, inplace=True)

# change 'time' values from string to datetime object
monarch_df['time'] = pd.to_datetime(monarch_df['time'])
```

```python
# milkweed butterfly csv files from iNaturalist
milkweed_2008_2021 = pd.read_csv('milkweed_2008-2021.csv')
milkweed_2021_2024 = pd.read_csv('milkweed_2021-2024.csv')
milkweed_df = pd.concat([milkweed_2008_2021, milkweed_2021_2024], ignore_index=True)

#cleaning milkweed data
# drop unecessary columns
columns_to_keep = ['observed_on', 'latitude', 'longitude']
milkweed_df = milkweed_df[columns_to_keep]

# drop null rows
milkweed_df = milkweed_df.dropna()

# rename observed_on column to time
milkweed_df.rename(columns={'observed_on': 'time'}, inplace=True)

# change 'time' values from string to datetime object
milkweed_df['time'] = pd.to_datetime(milkweed_df['time'])
```

## Temperature Data set

```python
#Temperature dataset from _________
#open NetCDF file and access 'temperature' variable
dataset = xr.open_dataset('Land_and_Ocean_LatLong1.nc')
temperature_data = dataset['temperature']

# convert DataArray to a pandas DataFrame
temperature_df = temperature_data.to_dataframe()

# reset the index to keep coordinates as columns
temperature_df.reset_index(inplace=True)
```

```python
#cleaning temperature data
# helper function to check if year is leap year
def is_leap(year):
    return (year % 4 == 0 and (year % 400 == 0 or year % 100 != 0))

# function to convert from fractional year to datetime object taking leap years into account
def convert_fractional_year(frac_year):
    if isinstance(frac_year, datetime):
        return frac_year
    year = int(frac_year)
    frac = frac_year - year
    if (year % 4):
        days_in_year = 366
    else:
        days_in_year = 365
    days_passed = int(days_in_year * frac)
    start_date = datetime(year, 1, 1)
    date = start_date + timedelta(days=days_passed)
    return date

# drop null rows
temperature_df = temperature_df.dropna()

# drop rows not in latitude-longitude range
min_lat = min(monarch_df['latitude'].min(), milkweed_df['latitude'].min())
max_lat = max(monarch_df['latitude'].max(), milkweed_df['latitude'].max())
min_long = min(monarch_df['longitude'].min(), milkweed_df['longitude'].min())
max_long = max(monarch_df['longitude'].max(), milkweed_df['longitude'].max())
temperature_df = temperature_df[(temperature_df['latitude'] >= min_lat) & (temperature_df['latitude'] <= max_lat)]
temperature_df = temperature_df[(temperature_df['longitude'] >= min_long) & (temperature_df['longitude'] <= max_long)]

# convert time from fractional year to YYYY-MM-DD format
temperature_df['time'] = temperature_df['time'].apply(convert_fractional_year)

# drop rows not in date range
start_date = min(monarch_df['time'].min(), milkweed_df['time'].min())
end_date = max(monarch_df['time'].max(), milkweed_df['time'].max())
temperature_df = temperature_df[(temperature_df['time'] >= start_date) & (temperature_df['time'] <= end_date)]

# change latitude and longitude value types to float64
temperature_df[['latitude', 'longitude']] = temperature_df[['latitude', 'longitude']].astype(float)
```

## Merging data

To create the final dataframe, we wanted to create a region column. Rather than make the region based off of state/province, we decided to choose the region with K-means clustering (note: the number of clusters that will be used is undecided and may be adjusted in the future). We then grouped the temperature, milkweed, and monarch data by month and region. We also created a new column which is the actual temperature in degrees Celsius, rather than the temperature measured as anomolies relative to the baseline period.

```python
# combine all latitude and longitude data from dataframes
coordinates = np.vstack((
    temperature_df[['latitude', 'longitude']].values,
    milkweed_df[['latitude', 'longitude']].values,
    monarch_df[['latitude', 'longitude']].values))

# perform K-means clustering
num_clusters = 50  # to be adjusted
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
kmeans.fit(coordinates)

# assign clusters as regions back to each dataframe
temperature_df['region'] = kmeans.predict(temperature_df[['latitude', 'longitude']].astype(float))
milkweed_df['region'] = kmeans.predict(milkweed_df[['latitude', 'longitude']])
monarch_df['region'] = kmeans.predict(monarch_df[['latitude', 'longitude']])
```

```python
# aggregate temperature data on a monthly basis by region
temperature_monthly_avg = temperature_df.groupby(['region', pd.Grouper(key='time', freq='M')]).temperature.mean().reset_index()

# aggregate milkweed presence on a monthly basis by region
milkweed_presence = milkweed_df.assign(presence=1).groupby(['region', pd.Grouper(key='time', freq='M')]).presence.max().reset_index()

# aggregate monarch counts on a monthly basis by region
monarch_monthly_counts = monarch_df.groupby(['region', pd.Grouper(key='time', freq='M')]).size().reset_index(name='monarch_count')

# merge aggregated temperature and milkweed presence on region and time
features_df = pd.merge(temperature_monthly_avg, milkweed_presence, on=['region', 'time'], how='outer').fillna({'presence': 0})

# merge the above with monarch counts to include monarch population data
final_df = pd.merge(features_df, monarch_monthly_counts, on=['region', 'time'], how='left').fillna({'monarch_count': 0})

# drop null rows
final_df = final_df.dropna()

# create new column called actual_temperature which is the temperature as an anomoly plus the baseline average
baseline_temperature = 14
final_df['actual_temperature'] = final_df['temperature'] + baseline_temperature
```
